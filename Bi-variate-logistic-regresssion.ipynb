```python
import random
def random_gen(s, e, n):
    randomlist = []
    for i in range(n):
        ri = random.uniform(s, e) # for radom float value
        rj = random.uniform(s, e)
        randomlist.append([ri, rj])
        #print(randomlist)
    return randomlist
```

```python
def sigmoid(z):
    sigmd = 1 / (1 + np.exp(-z))
    return sigmd
```

```python
# Iterations
epochs = 10000
lr = 0.0001 # learning_rate

m1, m2, b = 1, 1, 1
old_m1, old_m2, old_b0 = m1, m2, b

finall_loss_list = []
finall_m1_list = []
finall_m2_list = []
finall_b_list = []

for epoch in range(epochs):
    
    # linear Equation
    fcost = 0
    cost = 0
    loss_list = []
    n = len(x_train)
    
    for x1,y1 in zip(x_train,y_train): 
        
        #print(f"x1[0]: {x1[0]}, x1[1]: {x1[1]}, y1[0]: {y1[0]}, y1[1]: {y1[1]}")
        
        #z = old_m1*x1 + old_b0
        z = (old_m1*x1[0]) + (old_m2*x1[1]) + old_b0
        
        # Sigmoid
        a = sigmoid(z)
        #print(f"a : {a}")
        
        # Loss
        cost = -(y1*np.log(a) + (1-y1)*np.log(1-a))
        loss_list.append(cost)
        
        # Derivative of m and b
        dm1 = ((a - y1)* x1[0]) 
        dm2 = ((a - y1)* x1[1])
        db = (a - y1)
        #print(f"dm1: {dm1}, dm2: {dm2}, db: {db}")
        
        # update old_m1, old_m2, old_b
        old_m1 = old_m1 - (lr*dm1)
        old_m2 = old_m2 - (lr*dm2)
        old_b0 = old_b0 - (lr*db)
        #print(f"old_m1: {old_m1}, old_b2: {old_b2}")
        
    floss = sum(loss_list) / n
    print(f"iteration - {epoch} -> cost: {floss}, new-m1: {old_m1}, new-m2: {old_m2}, new-b: {old_b0}")  
    
    finall_loss_list.append(floss)
    finall_m1_list.append(old_m1)
    finall_m2_list.append(old_m2)
    finall_b_list.append(old_b0)        
```
